```python
load_dotenv()

app = FastAPI()

def initialize_components():
    try:
        loader = CSVLoader('shl_product_catalog.csv')
        data = loader.load()

        embedding_model = HuggingFaceEmbeddings(
            model_name="sentence-transformers/all-MiniLM-L6-v2"
        )

        DB_FAISS_PATH = "vectorstore/db_faiss"
        if os.path.exists(DB_FAISS_PATH):
            vectorstore = FAISS.load_local(
                DB_FAISS_PATH,
                embedding_model,
                allow_dangerous_deserialization=True
            )
        else:
            vectorstore = FAISS.from_documents(data, embedding_model)
            vectorstore.save_local(DB_FAISS_PATH)

        llm = HuggingFaceEndpoint(
            repo_id="mistralai/Mistral-7B-Instruct-v0.3",
            task="text-generation",
            temperature=0.5,
            max_length=512,
            api_key="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"
        )

        PROMPT_TEMPLATE = """
        You are an intelligent recommendation assistant that helps hiring managers choose the most relevant SHL individual assessments.

        Use only the information provided in the context to recommend SHL assessments based on the user's input.
        Do NOT hallucinate or assume any information that is not in the context.
        Give main focus on job role, test type and duration.

        Each recommendation must include:
        - Assessment Name
        - URL (must be from SHL product catalog)
        - Remote Testing Support (Yes/No)
        - Adaptive/IRT Support (Yes/No)
        - Duration
        - Test Type

        Context: {context}
        Question: {question}

        Recommend only 10 listings.
        Start your answer directly. No introductions or small talk.
        """

        prompt = PromptTemplate(
            template=PROMPT_TEMPLATE,
            input_variables=["context", "question"]
        )

        qa_chain = RetrievalQA.from_chain_type(
            llm=llm,
            chain_type="stuff",
            retriever=vectorstore.as_retriever(search_kwargs={'k': 3}),
            return_source_documents=True,
            chain_type_kwargs={'prompt': prompt}
        )

        return qa_chain

    except Exception as e:
        raise RuntimeError(f"Initialization failed: {str(e)}")

try:
    qa_chain = initialize_components()
except Exception as e:
    print(f"Failed to initialize: {e}")
    exit(1)

class QueryInput(BaseModel):
    query: str

class AssessmentRecommendation(BaseModel):
    name: str
    url: str
    remote_testing: str
    adaptive_support: str
    duration: str
    test_type: str

class RecommendationResponse(BaseModel):
    recommendations: List[AssessmentRecommendation]

@app.post("/recommend", response_model=RecommendationResponse)
async def get_recommendations(query: QueryInput):
    try:
        result = qa_chain.invoke({'query': query.query})

        recommendations = parse_llm_output(result["result"])

        return {"recommendations": recommendations}

    except Exception as e:
        raise HTTPException(
            status_code=500,
            detail=f"Recommendation failed: {str(e)}"
        )

def parse_llm_output(output: str) -> List[Dict]:
    """Parse the LLM's text output into structured recommendations"""
    return [{
        "name": "Sample Assessment",
        "url": "https://www.shl.com/example",
        "remote_testing": "Yes",
        "adaptive_support": "No",
        "duration": "30 minutes",
        "test_type": "Cognitive"
    }]
```
