```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import pandas as pd
import time

# Step 1: Open browser
driver = webdriver.Chrome()
driver.get("https://www.shl.com/solutions/products/product-catalog/")
time.sleep(5)  # Wait for the page to load

window_height = driver.execute_script("return window.innerHeight")
driver.execute_script(f"window.scrollBy(0, {window_height * 1.5});")
time.sleep(2)

rows = driver.find_elements(By.XPATH, "/html/body/main/div[3]/div/div/div/div[2]/div")

data = []  # List to store all rows

for i, row in enumerate(rows):
    tds = row.find_elements(By.TAG_NAME, "td")
    row_data = {
        "Product Name": "",
        "Link": "",
        "Keys": "",
        "Title": "",
        "Description": "",
        "Job": "",
        "Language": "",
        "Time": "",
        "Remote Testing":"",
        "Adaptive IRT":""
    }

    print(f"Row {i + 1}:")
    for j, td in enumerate(tds):
        if 'custom__table-heading__title' in td.get_attribute('class'):
            try:
                a_tag = td.find_element(By.TAG_NAME, 'a')
                link = a_tag.get_attribute('href')
                name = a_tag.text.strip()
                print(f"  Cell {j + 1} (Link): {link}")
                row_data["Product Name"] = name
                row_data["Link"] = link

                # Open link
                driver.execute_script("window.open(arguments[0]);", link)
                driver.switch_to.window(driver.window_handles[1])
                time.sleep(5)

                try:
                    title = driver.find_element(By.XPATH, "/html/body/main/div[2]/div/div[1]/h1").text.strip()
                    row_data["Title"] = title
                    print("    ‚Ü≥ Page Title:", title)
                except:
                    print("    ‚Ü≥ Could not extract title")

                try:
                    desc = driver.find_element(By.XPATH, "/html/body/main/div[2]/div/div[2]/div[1]/div/div[1]/p").text.strip()
                    row_data["Description"] = desc
                    print("    ‚Ü≥ Page desc:", desc)
                except:
                    print("    ‚Ü≥ Could not extract desc")

                try:
                    job = driver.find_element(By.XPATH, "/html/body/main/div[2]/div/div[2]/div[1]/div/div[2]/p").text.strip()
                    row_data["Job"] = job
                    print("    ‚Ü≥ Page job:", job)
                except:
                    print("    ‚Ü≥ Could not extract job")

                try:
                    lang = driver.find_element(By.XPATH, "/html/body/main/div[2]/div/div[2]/div[1]/div/div[3]/p").text.strip()
                    row_data["Language"] = lang
                    print("    ‚Ü≥ Page lang:", lang)
                except:
                    print("    ‚Ü≥ Could not extract lang")

                try:
                    timing = driver.find_element(By.XPATH, "/html/body/main/div[2]/div/div[2]/div[1]/div/div[4]/p").text.strip()
                    row_data["Time"] = timing
                    print("    ‚Ü≥ Page time:", timing)
                except:
                    print("    ‚Ü≥ Could not extract time")

                driver.close()
                driver.switch_to.window(driver.window_handles[0])
                time.sleep(1)

            except Exception as e:
                print(f"  Cell {j + 1}: Error - {e}")

        elif 'custom__table-heading__general product-catalogue__keys' in td.get_attribute('class'):
            spans = td.find_elements(By.CLASS_NAME, "product-catalogue__key")
            keys = [span.text.strip() for span in spans if span.text.strip()]
            row_data["Keys"] = ", ".join(keys)
            print(f"  Cell {j + 1} (Keys): {row_data['Keys']}")

        elif 'custom__table-heading__general' in td.get_attribute('class'):
            try:
                td.find_element(By.TAG_NAME, 'span')
                print(f"  Cell {j + 1}: yes")
            except:
                print(f"  Cell {j + 1}: no")

        else:
            print(f"  Cell {j + 1}: {td.text.strip()}")
        time.sleep(2)

    data.append(row_data)

# Export to CSV
df = pd.DataFrame(data)
df.to_csv("shl_product_catalog.csv", index=False)
print("\n‚úÖ CSV file 'shl_product_catalog.csv' created!")

# Optionally quit browser
# driver.quit()
```
